```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import datetime

"""
Récupération des rendements des actifs du DJIA
""" 



def get_rendements(stocks,début,fin): #date=yyyy-mm-dd en string
    prix=pdr.get_data_yahoo(stocks,start=début,end=fin)['Close'] 
    returns=prix.pct_change()
    logrdt=np.log(1+returns)*100
    logrdt=logrdt.iloc[1:]
    returns=returns.iloc[1:]
    prix=prix.iloc[1:]
    return  returns, prix, logrdt

stocks=['AAPL','AXP','BA', 'CAT', 'CSCO', 'CVX','DD','DIS','GE','HD','HPQ','IBM','INTC','JNJ','JPM','KO','MCD','MMM','MRK','MSFT','NKE','PFE','PG','RTX','T','TRV','UNH','VZ','WMT','XOM']
début='1995-12-29'
fin='2014-12-31'

output=get_rendements(stocks, début, fin)
rendements=output[0]
prix=output[1]
rdt_log=output[2]
```

```python
import numpy as np



from scipy.stats import norm


def VaR_HS(rdt,sample_size=250,var_quantile=0.01):
    """
    Historical simulation
    """
    
    # Calcul de la moyenne
    mu=np.mean(rdt)
    # Calcul de l'écart-type
    sigma=np.std(rdt)
    # Création d'un échantillon de moyenne mu et d'écart-type sigma
    sample=norm.rvs(loc=mu,scale=sigma, size=sample_size)
    # Tri de l'échantillon
    sample=np.sort(sample)
    # Calcul de la VaR
    VaR=sample[int(sample_size*var_quantile)-1]
    
    return VaR
```

```python
"""
Calcul de la VaR HS
"""

import numpy as np

VaR_HS_Asset=np.zeros((len(rdt_log),30))


for j in range(0,30):
    for i in range(250,len(rdt_log)):
        VaR_HS_Asset[i,j]=VaR_HS(rdt_log.iloc[i-250:i,j],250,var_quantile=0.01)
```
